```python
import pandas as pd

df = pd.read_csv("data/sessions.csv")

# parse datetimes
df['start'] = pd.to_datetime(df['start_time'], dayfirst=True)
df['end'] = pd.to_datetime(df['end_time'], dayfirst=True)

# create hourly timestamps
rng = pd.date_range(df['start'].min().floor('H'),
                    df['end'].max().ceil('H'),
                    freq='H')

# count concurrent
def count_concurrent(ts):
    return ((df['start'] <= ts) & (df['end'] > ts)).sum()

concurrent = pd.DataFrame({'timestamp': rng})
concurrent['concurrent_users'] = concurrent['timestamp'].apply(count_concurrent)

concurrent
```
