```python
import pandas as pd

# Sample DataFrame
data = {'Category': ['A', 'B', 'A', 'C', 'B', 'A', 'C'],
        'Value': [10, 20, 15, 25, None, 12, 30], "another" : [1, 23, 4, 55, 1, 3, 5,]}
df = pd.DataFrame(data)

# Group by 'Category' and count non-null 'Value's
grouped_counts = df.groupby('Category')['Value'].count()
print(grouped_counts)
```

```python
import datetime

td = datetime.timedelta(days=13, hours=5, seconds=45)

days_int = td.days
seconds_component_int = td.seconds # This returns only the seconds component within the current day/hour/minute
microseconds_component_int = td.microseconds

print(f"Timedelta: {td}")
print(f"Days: {days_int}")
print(f"Seconds component: {seconds_component_int}")
print(f"Microseconds component: {microseconds_component_int}")
```

```python
import pandas as pd

# Create a sample Series of datetime objects
dates = pd.Series(pd.to_datetime(['2024-01-01', '2024-02-15', '2024-11-30']))

# Convert the Series to strings with a specific format (e.g., 'YYYY-MM-DD')
string_dates = dates.dt.strftime('%Y-%m-%d')

print("Original Datetime Series:")
print(dates)
print("\nConverted String Series:")
print(string_dates)

# Example with a different format (e.g., 'Month Day, Year')
string_dates_formatted = dates.dt.strftime('%B %d, %Y')
print("\nConverted String Series (Formatted):")
print(string_dates_formatted)
```
